```python
#imports
import snscrape.modules.twitter as sntwitter
import pandas as pd
import time
```

```python
#function to scrape tweets

def get_tweets (year, number, search_term):
    # Setting variables to be used below
    maxTweets = number

    # Creating list to append tweet data to
    tweets_list = []

    # Using TwitterSearchScraper to scrape data and append tweets to list
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(f'{search_term} since:{year}-01-01 until:{year}-12-01').get_items()):
        if i>maxTweets:
            break
        tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
        
    return pd.DataFrame(tweets_list, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])

#Code adapted from https://github.com/MartinBeckUT/TwitterScraper/blob/master/snscrape/python-wrapper/snscrape-python-wrapper.ipynb
```

```python
#function to scrape tweets for specific months 

def pull_tweets (year,month,number,search_term):
    
    # Setting variables to be used below
    maxTweets = number

    # Creating list to append tweet data to
    tweets_list = []

    # Using TwitterSearchScraper to scrape data and append tweets to list
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(f'{search_term} since:{year}-{month}-01 until:{year}-{month}-31').get_items()):
        if i>maxTweets:
            break
        tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
        
        
    return pd.DataFrame(tweets_list, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
```
